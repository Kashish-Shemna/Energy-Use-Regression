**Treating the Data**

We are using 3 types of treatments:

1. Z-score Outlier Treatment + Standardization
2. IQR (Interquartile Range) Outlier Treatment (Raw scale)
3. Min-Max Normalization

We will be running the regression on all 3 to find the better model



IQR Method

```python
# Make a fresh copy of the original dataframe
df_iqr = df.copy()

# Identify numeric columns except 'Lighting_Use_Wh'
iqr_cols = df_iqr.select_dtypes(include=[np.number]).columns.drop('Lighting_Use_Wh')

# Apply IQR-based capping
for col in iqr_cols:
    Q1 = df_iqr[col].quantile(0.25)
    Q3 = df_iqr[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip values outside IQR bounds
    df_iqr[col] = np.where(df_iqr[col] < lower_bound, lower_bound, df_iqr[col])
    df_iqr[col] = np.where(df_iqr[col] > upper_bound, upper_bound, df_iqr[col])
```

Z-Score Method

```python
from sklearn.preprocessing import StandardScaler

# Create a copy
df_clean = df.copy()

# Identify numeric columns and exclude 'Lighting_Use_Wh'
cols_to_scale = df_clean.select_dtypes(include=[np.number]).columns.drop('Lighting_Use_Wh')

# Apply StandardScaler to selected columns
scaler = StandardScaler()
df_scaled = df_clean.copy()
df_scaled[cols_to_scale] = scaler.fit_transform(df_clean[cols_to_scale])
```

```python
from scipy.stats import zscore
import numpy as np

# Copy to preserve scaled data
df_no_outliers = df_scaled.copy()

# Loop through all numeric columns except Lighting_Use_Wh
for col in cols_to_scale:
    col_z = zscore(df_no_outliers[col])
    outliers = np.abs(col_z) > 3  # Z-score threshold

    # Clip outliers to boundary values
    upper = df_no_outliers.loc[~outliers, col].max()
    lower = df_no_outliers.loc[~outliers, col].min()

    df_no_outliers.loc[col_z > 3, col] = upper
    df_no_outliers.loc[col_z < -3, col] = lower
```

Removing highly collinear variables

```python
import numpy as np

# Function to drop highly correlated features
def drop_highly_correlated(df, threshold=0.85):
    df_corr = df.drop(columns=['Timestamp'])  # drop non-feature column
    corr_matrix = df_corr.corr().abs()

    # Get upper triangle of correlation matrix
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

    # Identify columns to drop
    to_drop = [col for col in upper.columns if any(upper[col] > threshold)]

    # Drop from dataframe
    reduced_df = df.drop(columns=to_drop)
    return reduced_df, to_drop

# Apply to all 3 datasets
df_no_outliers_reduced, dropped_zscore = drop_highly_correlated(df_no_outliers)
df_iqr_reduced, dropped_iqr = drop_highly_correlated(df_iqr)
df_minmax_reduced, dropped_minmax = drop_highly_correlated(df_minmax)

# Print dropped columns
print("\n📌 Dropped from Z-score dataset:", dropped_zscore)
print("\n📌 Dropped from IQR dataset:", dropped_iqr)
print("\n📌 Dropped from Min-Max dataset:", dropped_minmax)
```

OLS Regression for all 3 methods

```python
import statsmodels.api as sm

# Define a helper function
def run_multiple_regression(df, label):
    print(f"\n📊 Multiple Regression Summary for: {label}")

    # Sort by timestamp
    df = df.sort_values('Timestamp')

    # Prepare features and target
    X = df.drop(columns=['Timestamp', 'Energy_Use_Wh'])
    y = df['Energy_Use_Wh']

    # Add intercept
    X = sm.add_constant(X)

    # Fit the OLS model
    model = sm.OLS(y, X).fit()

    # Display summary
    print(model.summary())

# Run on all three cleaned (reduced) datasets
run_multiple_regression(df_no_outliers_reduced, "Z-score + StandardScaler (Reduced)")
run_multiple_regression(df_iqr_reduced, "IQR Treatment (Reduced)")
run_multiple_regression(df_minmax_reduced, "Min-Max Scaled (Reduced)")
```

Running Regression with combinations of high VIF and insignificant variables

```python
from statsmodels.api import OLS, add_constant
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# Function to calculate VIFs
def compute_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Main feature selection function based on VIF and p-value
def vif_pval_selection(df, target_col='Energy_Use_Wh', vif_thresh=10, pval_thresh=0.05):
    df_copy = df.copy()
    df_copy['Timestamp'] = pd.to_datetime(df_copy['Timestamp'], dayfirst=True)
    df_copy = df_copy.sort_values('Timestamp')

    # Separate features and target
    X = df_copy.drop(columns=['Timestamp', target_col])
    y = df_copy[target_col]

    # Add intercept
    X_const = add_constant(X)

    iteration = 0
    while True:
        iteration += 1
        model = OLS(y, X_const).fit()

        # Get p-values and VIFs
        pvalues = model.pvalues
        vif = compute_vif(X_const)

        # Drop by VIF if needed
        high_vif = vif[(vif["VIF"] > vif_thresh) & (vif["Feature"] != 'const')]
        # Drop by p-value if needed
        high_pval = pvalues[(pvalues > pval_thresh) & (pvalues.index != 'const')]

        if not high_vif.empty:
            to_drop = high_vif.sort_values(by="VIF", ascending=False).iloc[0]["Feature"]
            print(f"[Iteration {iteration}] Dropping by VIF > {vif_thresh}: {to_drop}")
        elif not high_pval.empty:
            to_drop = high_pval.sort_values(ascending=False).index[0]
            print(f"[Iteration {iteration}] Dropping by p-value > {pval_thresh}: {to_drop}")
        else:
            break

        # Drop and update X
        X_const = X_const.drop(columns=to_drop)

    print("\n✅ Final model summary after VIF and p-value filtering:")
    final_model = OLS(y, X_const).fit()
    print(final_model.summary())

    return X_const.columns.tolist(), final_model

# Example usage on the IQR dataset:
# selected_features, final_model = vif_pval_selection(df_iqr_reduced)
```
